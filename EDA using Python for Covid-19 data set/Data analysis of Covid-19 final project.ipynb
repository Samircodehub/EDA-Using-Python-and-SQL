# Import Libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns
import numpy as np
```

# Loading Dataset

```python
df=pd.read_csv(r"C:\Users\Samir\Downloads\covid_19_india.csv",parse_dates=["Date"],dayfirst=True)
df
```

## Removing unnecessary columns

```python
df = df.drop(['ConfirmedIndianNational', 'ConfirmedForeignNational', 'Time'], axis = 1)
df.head()
```

```python
df = df.drop(['Sno'], axis = 1)
df.head()
```

## Reanaming 'State/UnionTerritory' to 'state'

```python
df.rename(columns = {'State/UnionTerritory':'state'}, inplace = True)
```

```python
df
```

## Cheaking for null value

```python
df.isnull().sum()
```

```python
today=df[df.Date=="2021-08-11"]
today
```

```python
#sorting data with respect to confirmed cases
max_confirmed_cases=today.sort_values(by="Confirmed",ascending=False)
```

```python
max_confirmed_cases.head(10)
```

```python
# Extract top ten states with comfirmed
top_states_confirmed=max_confirmed_cases[0:10]
top_states_confirmed
```

## Visualizing number of confirmed cases with respect to states (Top 10 states)

```python
sns.set(rc={"figure.figsize":[15,8]})
sns.barplot(x="state",y="Confirmed",data=top_states_confirmed,hue="state")
plt.show()
```

```python
max_deaths_cases=today.sort_values(by="Deaths",ascending=False)
max_deaths_cases.head(10)
```

```python
top_states_deaths=max_deaths_cases[0:10]
top_states_deaths
```

## Visualizing number of deaths with respect to states (Top 10 states)

```python
sns.set(rc={"figure.figsize":[15,8]})
sns.barplot(x="state",y="Deaths",data=top_states_deaths,hue="state")
plt.show()
```

```python
max_cured_cases=today.sort_values(by="Cured",ascending=False)
max_cured_cases.head(10)
```

```python
top_states_cured=max_cured_cases[0:10]
top_states_cured
```

## Visualizing the number of cured cases with respect to state

```python
sns.set(rc={"figure.figsize":[15,8]})
sns.barplot(x="state",y="Cured",data=top_states_cured,hue="state")
plt.show()
```

# Analysis for Maharastra state

```python
maha=df[df.state=="Maharashtra"]
maha
```

```python
maha.head()
```

```python
maha.tail()
```

## Visualizing confirmed cases trend with date in maharastra


```python
sns.set(rc={"figure.figsize":(20,8)})
sns.lineplot(x="Date",y="Confirmed",data=maha,color="green")
plt.show()
```

## Visualizing death cases trend with date in maharastra


```python

sns.set(rc={"figure.figsize":(15,8)})
sns.lineplot(x="Date",y="Deaths",data=maha,color="red")
plt.show()
```

## Visualizing Cured cases trend with date in maharastra


```python
sns.set(rc={"figure.figsize":(15,8)})
sns.lineplot(x="Date",y="Cured",data=maha,color="red")
plt.show()
```

```python
sns.set(rc={"figure.figsize":(15,8)})
sns.relplot(x="Date",y="Cured",data=maha,hue= 'Cured')
plt.title("No. of Cured Trend")
plt.xticks(rotation=45)
plt.show()
```
